```python
################################################################################################
# Python coder : sjkim (kim.sajoong@gmail.com)
# last update  : 2021-Feb-08
#----------------------------------------------------------------------------------------------#
from bing_image_downloader import downloader
from tqdm.notebook import tqdm, trange
import time, random, os, sys
from glob import glob
import pyautogui
#----------------------------------------------------------------------------------------------#
try:  
    search_image_name = pyautogui.prompt(title='Bing Image Search & Download',
                                         text='빙 이미지 가져오기', 
                                         default='이미지 이름을 입력하세요.')
    print(f"이미지 검색 이름 : [{search_image_name}]")
    if search_image_name == None or search_image_name == "이미지 이름을 입력하세요.":
        raise Exception("[이미지 검색어 오류]")
        
except Exception as e:
    print(e, "검색할 이미지가 없습니다.")
    sys.exit()
#----------------------------------------------------------------------------------------------#    
if len(search_image_name.split()) > 1:  # 아름다운 배경화면 >>> 아름다운_배경화면 으로 수정하기 
    search_image_dir = search_image_name.split()
    search_image_dir = "_".join(search_image_dir)
else:
    search_image_dir = search_image_name
#----------------------------------------------------------------------------------------------#
try:
    number = pyautogui.prompt(title   = 'Bing Image Search & Download',  
                              text    = f"다운로드할 이미지 갯수를 입력하세요.",
                              default = "10")
    number = int(number)   
except Exception as e:
    number = 10           # 디폴트 값을 10으로 설정
    
print(f"다운로드할 이미지 갯수는 {number}개 입니다.")    
#----------------------------------------------------------------------------------------------# 
downloader.download(search_image_name, 
                    limit             = number, 
                    adult_filter_off  = True, 
                    force_replace     = False,
                    timeout           = 60)
#----------------------------------------------------------------------------------------------#
print(os.getcwd())
dst_dir = os.path.join("dataset", search_image_dir)

if not os.path.isdir(dst_dir):
    src_dir = os.path.join("dataset", search_image_name)
    os.rename(src_dir, dst_dir)
else:
    pass

path = os.path.join(dst_dir, "*.*")
file_list = glob(path)

for file in file_list:
    src_file_name    = file.split("\\")[-1]
    src_file_number  = int(src_file_name.split("_")[-1].split(".")[0])
    dst_file_ext     = src_file_name.split("_")[-1].split(".")[-1]
    dst_file_name    = os.path.join(dst_dir, 
                        f"{search_image_dir}_{src_file_number:03d}.{dst_file_ext}")
    os.rename(file, dst_file_name)
    print(dst_file_name)
#----------------------------------------------------------------------------------------------#
```
